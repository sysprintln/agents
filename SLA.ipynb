```python
# -----------------------------------
# SLA MISMATCH ESTIMATION ENGINE
# -----------------------------------

# 1. PROCESS CONFIGURATION (calculation ratio)
process_timing = {
    "a": 0.2,
    "b": 0.1,
    "c": 0.15   # example process
}

# 2. DATA VOLUME (in GB)
data_volume = {
    "a": 1.5,
    "b": 2.5,
    "c": 1.0
}

# 3. SLA METRICS (in minutes)
sla_metrics = {
    "a": 60,   # 1 hour
    "b": 30,
    "c": 45
}

# Default SLA (if nothing configured)
default_sla = 60


# -----------------------------------
# FUNCTION: Calculate runtime of process
# -----------------------------------
def calculate_process_time(process, dv, ratio):
    """
    Calculates time required for a process
    Formula: time = data_volume * calculation_ratio * 60 (minutes)
    """
    time = dv * ratio * 60
    return time


# -----------------------------------
# FUNCTION: Estimate total runtime
# -----------------------------------
def estimate_total_runtime(process_timing, data_volume):
    total_runtime = 0
    runtime_breakdown = {}

    for process in process_timing:
        dv = data_volume.get(process, 0)
        ratio = process_timing[process]

        time = calculate_process_time(process, dv, ratio)

        runtime_breakdown[process] = time
        total_runtime += time

    return -]\, total_runtime


# -----------------------------------
# FUNCTION: Get Applicable SLA
# -----------------------------------
def get_applicable_sla(sla_metrics):
    """
    Choose strict SLA (minimum SLA)
    """
    if sla_metrics:
        return min(sla_metrics.values())
    return default_sla


# -----------------------------------
# FUNCTION: Check SLA breach
# -----------------------------------
def is_sla_breached(total_runtime, sla):
    return total_runtime > sla


# -----------------------------------
# FUNCTION: RCA Trigger
# -----------------------------------
def trigger_rca(runtime, sla):
    print("\n🚨 ALERT: SLA BREACH PREDICTED!")
    print(f"Estimated Runtime: {runtime:.2f} mins")
    print(f"SLA: {sla} mins")
    print("👉 Triggering RCA (Root Cause Analysis) Agent...")


# -----------------------------------
# MAIN EXECUTION
# -----------------------------------
def main():

    print("📊 SLA Prediction Engine Running...\n")

    # Step 1: Calculate runtime
    runtime_breakdown, total_runtime = estimate_total_runtime(process_timing, data_volume)

    # Step 2: Print process details
    print("🔹 Process-wise Runtime:")
    for process, time in runtime_breakdown.items():
        print(f"Process {process}: {time:.2f} mins")

    # Step 3: Total runtime
    print(f"\n⏱ Total Estimated Runtime: {total_runtime:.2f} mins")

    # Step 4: Get SLA
    sla = get_applicable_sla(sla_metrics)
    print(f"🎯 Applicable SLA: {sla} mins")

    # Step 5: SLA check
    if is_sla_breached(total_runtime, sla):
        trigger_rca(total_runtime, sla)
    else:
        print("\n✅ SLA will be met successfully")


# Run program
if __name__ == "__main__":
    main()
```
